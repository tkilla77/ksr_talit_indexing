```python
def count_letters(filename):
    letters = {}
    with open(filename, encoding="utf-8") as data:
        for line in data:
            for letter in line.upper():
                if letter in letters:
                    letters[letter] += 1
                else:
                    letters[letter] = 1
    
    # total = sum(letters.values())
    # letters = {k: v / total for k, v in letters.items()}

    letters = dict(sorted(letters.items(), key=lambda item: item[1], reverse=True))
    return letters
```

```python
wpde = count_letters('data/countries_de.csv')
bible_de = count_letters('data/Bibel_Martin_Luther_1912_utf8.txt')
```

```python
import matplotlib.pyplot as plt
import string

def plot_dict(d, label):
    x = list(string.ascii_uppercase)
    total = sum(d.values())
    y = [d[key]/total for key in x]

    plt.plot(x, y, label=label)
    plt.legend()

plot_dict(wpde, "wikipedia de")
plot_dict(bible_de, "bible de")
```

```python
def count_words(filename):
    words = {}
    with open(filename, encoding="utf-8") as data:
        for line in data:
            for word in line.split():
                word = word.strip(",.;:!?/=(){}[]").upper()
                if word in words:
                    words[word] += 1
                else:
                    words[word] = 1
    
    # total = sum(letters.values())
    # letters = {k: v / total for k, v in letters.items()}

    words = dict(sorted(words.items(), key=lambda item: item[1], reverse=True))
    return words
```

```python
wpde_words = count_words('data/countries_de.csv')
list(wpde_words.items())[:20]
```

```python
biblede_words = count_words('data/Bibel_Martin_Luther_1912_utf8.txt')
list(biblede_words.items())[:20]
```

```python
def find_min(l, start):
    # Wir starten mit dem ersten Element ab start:
    minIndex = start
    # Jetzt schauen wir jedes weitere Element an, und merken uns, wenn wir ein kleineres finden:
    for index in range(start + 1, len(l)):
        if l[index] < l[minIndex]:
            minIndex = index
    return minIndex   
 
def topn(list, n):
    for index in range(n):
        min_index = find_min(list, index)
        list[index], list[min_index] = list[min_index], list[index] # swap
    return list[:n]
 
example =[12, 23, 15, 20, 13, 7]
print(topn(example, 2))
print(example)
```
